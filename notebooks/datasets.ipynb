```python
from survae.data import *
from survae.plot import *
```

```python
def plot_dataset_histogram(dataset, ax, conditional):
    """Plot a dataset as a 2D histogram with an offset to the bounds."""
    n = SAMPLE_COUNT
    
    data, labels = dataset(n)

    if conditional:
        unique_labels = np.unique(labels)

        for label in unique_labels:
            label_indices = np.where(labels == label)[0]
            ax.scatter(
                data[label_indices, 0], data[label_indices, 1],
                label=f'Label {label}', alpha=1000 / n
            )
        
    else:
        ax.hist2d(data[:, 0], data[:, 1], bins=BINS)
        
    ax.set_title(dataset.get_name())

def plot_datasets(datasets, labels=False):
    # Calculate the grid size for subplots
    num_datasets = len(datasets)
    num_cols = 3  # Number of columns in the grid
    num_rows = (num_datasets + num_cols - 1) // num_cols  # Calculate number of rows

    # Create subplots
    fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))

    # Plot each dataset
    for i, dataset in enumerate(datasets):
        row = i // num_cols
        col = i % num_cols
        plot_dataset_histogram(dataset, axs[row, col], labels)

    # Adjust layout
    plt.tight_layout()
    plt.show()
```

```python
# Plotting the datasets without modifications
plot_datasets(TWO_D_DATASETS())
```

```python
# Plotting conditional datasets without modifications
plot_datasets(TWO_D_DATASETS(), labels=True)
```
