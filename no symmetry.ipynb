```python
import torch
import matplotlib.pyplot as plt
```

Since the q-value has the same dimension as the data, it could also be interesting if twodimensional data only has one symmetry axis. In this case, the following dataset would be enough:

```python
def gradient_dataset(n: int, r: float = 1.0):
    x_axis = r * torch.rand(n)
    y_axis = r * torch.rand(n) ** (1/2)

    return torch.stack((x_axis, y_axis), dim=1)
```

```python
X = gradient_dataset(1_000_000, r=10)
```

```python
plt.figure(figsize=(4, 4))

plt.hist2d(X[:, 0], X[:, 1], bins=100)

plt.axis('equal')
plt.tight_layout()
plt.show()
```

If that turns out to be boring, there's a chance the following dataset might be more exciting. I think in theory, it has no perfect symmetry axis, although tbh I'm not sure what the SurVAE will count as a symmetry axis.

```python
def cool_dataset(n: int, r: float = 1.0):
    # The exponents were hand-picked such that the result looks "unsymmetric".
    # Their exact values shouldn't influence whether or not the model finds a symmetry axis.
    x_axis = r * torch.rand(n) ** (1/2.4)
    y_axis = r * torch.rand(n) ** (1/1.4)

    return torch.stack((x_axis, y_axis), dim=1)
```

```python
X = cool_dataset(1_000_000, r=5.3)
```

```python
plt.figure(figsize=(4, 4))

plt.hist2d(X[:, 0], X[:, 1], bins=100)

plt.axis('equal')
plt.tight_layout()
plt.show()
```
